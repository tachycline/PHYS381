```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
def projdrag(r, t, b, c, m):
    x,y,vx,vy = r
    
    vsq = vx**2 + vy**2
    vmag = np.sqrt(vsq)
    
    dx = vx
    dy = vy
    
    dvx = -b/m*vx - c*vsq*vx/vmag
    dvy = -9.8 -b/m*vy - c*vsq*vy/vmag
    return np.array([dx,dy,dvx,dvy])
```

```python
t = np.linspace(0,3.25,1000)
r0 = (0,0,100,100)
output = odeint(projdrag, r0, t, args=(0.1, 0.1, 1.0))
```
