```python
def calc_median(filename):
    values_dict = dict()
    df = pd.read_csv(filename)
    initial_energy = df['E'].iloc[0]
    final_energy = df['E'].iloc[-1]
    df['T/E'] = df['T']/df['E']
    energy_ratio = initial_energy/final_energy

    description = df.describe()
    median1 = description['thetadot_0(t)'].loc['50%']
    median2 = description['thetadot_1(t)'].loc['50%']
    median3 = description['theta_0(t)'].loc['50%']
    median4 = description['theta_1(t)'].loc['50%']
    median5 = description['p_0(t)'].loc['50%']
    median6 = description['p_1(t)'].loc['50%']
   # values_dict['p0med = medians'] = ('p_0(t)')
    #values_dict['p1med = medians'] = ('p_1(t)')
    description = df.describe()
    TE = description['T/E'].loc['50%']
    medians = df.apply(np.median)
    return energy_ratio, median1, median2, median3, median4, median5, median6, TE
    
```

```python
csvfiles = glob.glob("/home/cmckay/clusters/CQM/*.csv")
```

```python
energy_ratios = []

for thefile in csvfiles[0:640]:
    energy_ratios.append(calc_median(thefile))
```
