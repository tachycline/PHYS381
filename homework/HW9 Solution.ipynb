```python
TE_1 =df['0'].values
```

```python
TE_2 = TE_1 -1
```

```python
NOUTLIERS  = []
OUTLIERS = []
for number in TE_2:
    if -0.0000060 < number < 0.000010:
        NOUTLIERS.append(number)
    elif number <= -.000006:
        number = -.0000063
        OUTLIERS.append(number)
    else:
        number =.0000107
        OUTLIERS.append(number)
```

```python
ax = plt.hist(NOUTLIERS, bins = 50, rwidth = 0.80)
bx = plt.hist(OUTLIERS,bins = 50,  rwidth = 0.80, facecolor = 'pink')
plt.text(0.000006,53,"N = 585", alpha = 1, size = 15)
plt.text(0.000005,45,"    Red bars are outliers\nwith deviation > 1.0 *10^-6 ", alpha = 1)
plt.title("Relative deviation from energy conservation")
plt.xlabel("E_i/E_f-1")
plt.ylabel("number of runs")
fig = plt.gcf()
fig.set_figheight(10)
fig.set_figwidth(15)
plt.xlim(-0.0000065,0.000011)
```
